```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
t = pd.read_csv("Titanic.csv")
t.info()
```

```python
t.head()
```

```python
plt.figure(figsize=(15,10))
sns.countplot(x = 'Survived',data = t)
plt.show()
```

```python
plt.figure(figsize=(15,10))
sns.countplot(x = 'Sex',data = t)
plt.show()
```

```python
survived = t["Survived"] .value_counts()

sns.barplot(x=survived.index, y=survived.values)

plt.title("Number of Survivors by Sex")
plt.xlabel("Sex")
plt.ylabel("Number of Survivors")
plt.show()
```

```python
sns.regplot(x=t["Sex"], y=t["Survived"])

plt.xlabel("Sex")
plt.ylabel("Survived")
plt.show()
```

```python
sns.histplot(x='Age',data =t )
```

```python
t["Age"].mean()
```

```python
sns.boxplot(x='Pclass', y='Age', hue='Survived', data=t)
plt.title('Age Distribution by Survival and Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Age')
plt.legend(title='Survived', loc='upper right')
plt.show()
```

```python
sns.countplot(x='Pclass', hue='Survived', data=t)
plt.title('Survival Count by Passenger Class and Gender')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.legend(title='Survived', loc='upper right')
plt.show()
```

```python
sns.heatmap(t.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
sns.barplot(x='Embarked', y='Survived', data=t)
plt.title('Survival Rate by Embarkation Port')
plt.xlabel('Embarkation Port')
plt.ylabel('Survival Rate')
plt.show()
```
